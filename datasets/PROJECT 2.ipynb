```python
nationality_clean = nationality.loc[(nationality.Year == 2017),['Year', 'District Name','Nationality','Number']]
```

```python
nationality_clean.head()
```

```python
###parte de analisis = primero group by disctrict and nationality
temp_df = nationality_clean.groupby(['District Name', 'Nationality']).sum().drop('Year', axis=1)
```

```python
temp_df.head()
#vemos que ahora tenemos todos los disctricts y luego todas las nacionalidades ordenadas por orden alfabetico
```

```python
#####ahora o bien definimos una funcion o hacemos una lambda function que nos coja cada 
#valor de numeros ynos lo ordene de mayoor a menor , ponemos ascending False pq lo queremos 
#descending.

def sort_district(district):
    return district.sort_values(['Number'], ascending=False)
```

```python
lambda district: district.sort_values(['Number'], ascending=False)
```

```python
####ahora como ya hemos agrupado por Disctrict name y nationality, 
##vemos que está en negrita, eso significa que para pandas es un único index
## necesitamos decirle que nos haga un group by solo por el level 0 del index (por disctrict)
##group_keys parameter in groupby comes handy during apply operations that creates an additional 
##index column corresponding to the grouped columns[group_keys=True] , es decir, nos crea otro indice y 
##en este caso no nos interesa, por lo tanto, False


temp_df.groupby(level=0, group_keys=False).sort_values()
```

```python
## a temp_df2 hazme un group by por disctrict y aplicame la lambda function
temp_df2 = temp_df.groupby(level=0, group_keys=False).apply(lambda district: district.sort_values(['Number'], ascending=False))
```
